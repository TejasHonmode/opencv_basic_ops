```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Setting the arg to 0 implies that the first webcam in your system is being
#engaged(if you have more than 1)
cap = cv2.VideoCapture(0)

#You can also access some of the features of this video using 
#cap.get(propId) method where propId is a number from 0 to 18. 
#Each number denotes a property of the video 
#(if it is applicable to that video) and full details can be seen here: Property Identifier.
#Some of these values can be modified using cap.set(propId, value). 
#Value is the new value you want.

#For example, I can check the frame width and height by cap.get(3) and 
#cap.get(4). It gives me 640x480 by default. 
#But I want to modify it to 320x240. 
#Just use ret = cap.set(3,320) and ret = cap.set(4,240).

while True:
    #ret - gives a true or false
    #frame - the feed you will see
    #cap.read() returns a bool (True/False). If frame is read correctly, 
    #it will be True. 
    #So you can check end of the video by checking this return value.
    ret, frame = cap.read()
    cv2.imshow('frame',frame)
    #if the key = 1 and q
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
        
cap.release()
cv2.destroyAllWindows()
```

```python
#Grayscaling our feed
cap = cv2.VideoCapture(0)

while True:
    #ret - gives a true or false
    #frame - the feed you will see
    #cap.read() returns a bool (True/False). If frame is read correctly, 
    #it will be True. 
    #So you can check end of the video by checking this return value.
    ret, frame = cap.read()
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    cv2.imshow('frame',frame)
    cv2.imshow('gray',gray)
    #if the key = 1 and q
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
        
cap.release()
cv2.destroyAllWindows()
```

```python
#To output(save) the video

cap = cv2.VideoCapture(0)
#cv2.VideoWriter_the_codec_used(fourcc here)
fourcc = cv2.VideoWriter_fourcc(*'XVID')
#cv2.VideoWriter(name_you_want_for_file,codec,no_of_frames,size_of_feed)
out = cv2.VideoWriter('output.avi',fourcc,20.0,(640,480))
while True:
    #ret - gives a true or false
    #frame - the feed you will see
    #cap.read() returns a bool (True/False). If frame is read correctly, 
    #it will be True. 
    #So you can check end of the video by checking this return value.
    ret, frame = cap.read()
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    out.write(frame)
    cv2.imshow('frame',frame)
    cv2.imshow('gray',gray)
    #if the key = 1 and q
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
        
cap.release()
cv2.destroyAllWindows()
```

```python
#Drawing and writing on image
```

```python
import numpy as np
import cv2
```

```python
#Reading the img. in colour
img = cv2.imread('watch.jpg',cv2.IMREAD_COLOR)
```

```python
#To draw a line on the image

#cv2.line(image,st_pt,end_pt,bgr,linewidth)
cv2.line(img,(0,0),(150,150),(255,255,255),15)
```

```python
cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#To draw rectaingle

#cv2.rectangle(img,top-left,bottom-right,bgr,linewidth)
cv2.rectangle(img,(15,25),(200,150),(0,255,0),5)

cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#To draw a circle

#cv2.circle(image,centre,radius,bgr,linewidth)
#If linewidth == -1 it will fill in the circle
cv2.circle(img,(100,63),55,(0,0,255),-1)

cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#To draw an ellipse
#cv2.ellipse(image,centre,(major_ax,minor_ax),ang._of_rotation of ellipse
#in anti-clock,start_ang.,end_angle(measured in clockwise direction from major axis),
#bgr,linewidth)
cv2.ellipse(img,(169,87),(100,50),0,0,180,(255,0,0),-1)

cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#To draw a polygon

pts = np.array([[10,5],[20,30],[70,20],[50,10]],np.int32)
#(ROWS,1,2)
#We have to reshape it
pts = pts.reshape((-1,1,2))
cv2.polylines(img,[pts],True,(0,255,255),3)

cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#To write on the image

font = cv2.FONT_HERSHEY_SIMPLEX

#cv2.putText(img,Text,st_pt,font,size,bgr,thickness,anti-aliasing)
cv2.putText(img,'OpenCV Tuts!',(0,130),font,1,(200,255,255),2,cv2.LINE_AA)

cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#These 2 imgs have to be of the same size
img1 = cv2.imread('3D-Matplotlib.png')
img2 = cv2.imread('mainsvmimage.png')
```

```python
plt.imshow(img1)
```

```python
plt.imshow(img2)
```

```python
add  = img1 + img2
```

```python
cv2.imshow('add',add)
cv2.waitKey(0)
cv2.destroyAllWindows()

plt.imshow(add)
```

```python
#The built-in addition operation 

add = cv2.add(img1,img2)

cv2.imshow('add',add)
cv2.waitKey(0)
cv2.destroyAllWindows()

plt.imshow(add)
```

```python
#This is all white bcoz cv.add() literally adds the pixels values of both
#the images
```

```python
#image blending using weights

weighted = cv2.addWeighted(img1,0.6,img2,0.4,0)

cv2.imshow('weighted',weighted)
cv2.waitKey(0)
cv2.destroyAllWindows()

plt.imshow(weighted)
```
